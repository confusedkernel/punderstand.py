```python
# Step 1: Import Required Libraries
import joblib
from visualization import visualize_prediction  # Import visualization function
from preprocessing import preprocess_text  # Ensure preprocessing matches training pipeline

# Step 2: Load the Model and Vectorizer
model_path = "knn_model.pkl"  # Change to your actual model path
pipeline = joblib.load(model_path)

# Step 3: Define the Prediction Function
def predict_with_visualization(pipeline, text):
    """
    Predicts whether the input text contains a pun and visualizes the result.
    """
    # Preprocess and predict
    processed_text = preprocess_text(text)  # Ensure preprocessing matches the training pipeline
    prediction = pipeline.predict([processed_text])[0]
    probability = (
        pipeline.predict_proba([processed_text])[0]
        if hasattr(pipeline, "predict_proba")
        else None
    )

    # Display and visualize
    if probability is not None:
        print(f"Prediction: {'Pun' if prediction == 1 else 'No Pun'} (Confidence: {max(probability) * 100:.2f}%)")
    else:
        print(f"Prediction: {'Pun' if prediction == 1 else 'No Pun'}")

    visualize_prediction(text, prediction, probability)

# Step 4: Test with an Example Input
input_text = "This is a pun example!"  # Replace with your own input
predict_with_visualization(pipeline, input_text)
```
